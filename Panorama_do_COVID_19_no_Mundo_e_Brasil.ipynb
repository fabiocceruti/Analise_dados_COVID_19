# Panorama do COVID-19 no Mundo e Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

### **Q1. Como está a evolução mundial da COVID-19?**

Desde do primeiro caso de contaminação por COVID-19 em Wuhan na China, tivemos que nos adaptar e adotar medidas de segurança, visando conter a propagação e os estragos causados pelo vírus.

Portanto, vamos avaliar como o número de casos e mortes relacionadas com o COVID-19 tem evoluído mundialmente. Para esta análise, será utilizado a média móvel (7 dias) de novos casos e mortes, pois facilita a avaliação em relação a evolução dos números.

```
#Definindo um dataframe apenas de dados relacionados com dados mundiais
df_sel = df_world.loc[df_world['location'] == 'World'].copy()

#Definindo os parâmetros de style para o matplotlib
rc_params = {'axes.edgecolor':'#787878', 
             'axes.titlecolor':'#787878', 
             'axes.labelcolor': '#787878', 
             'axes.spines.top':False, 
             'axes.spines.right': False, 
             'xtick.color': '#787878', 
             'ytick.color': '#787878',
             'axes.titleweight': 'bold',
             'axes.titlesize': 12
             }

#Aplicando os parâmetros no matplotlib/seaborn
with plt.rc_context(rc_params):

  #instanciando o axes e figure do matplotlib
  fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (22,8))

  #Construindo o gráfico de linha com a média móvel de novos casos 7 dias
  sns.lineplot(
                x = 'date',
                y = 'new_cases_smoothed',
                ci = None,
                ax = ax[0],
                data = df_sel  
              )
  
  #Colocando o título no gráfico
  ax[0].set_title("Média móvel (7 dias) mundial de novos casos")

  #Construindo o gráfico de linha com a média móvel de novas mortes 7 dias
  sns.lineplot(
                x = 'date',
                y = 'new_deaths_smoothed',
                ci = None,
                ax = ax[1],
                data = df_sel,
                color = "#621940" 
              )
  
  #Colocando o título no gráfico
  ax[1].set_title("Média móvel (7 dias) mundial de novas mortes")

  #Exibindo o gráfico
  plt.show()
```

As evoluções de novos casos e mortes podem estar associadas à:
* Medidas de controle flexíveis para países mais populosos, causando maior impacto na curva de casos e mortes mundiais;
* No inicio do segundo trimestre de 2020, a falta de conhecimento de como proceder em relação ao tratamento das pessoas com COVID-19 somados a falta de leitos hospitalares para atender todos pacientes podem ter relação o número expressivo de mortes. É importante destacar que quando os dois gráficos temporais são comparados, é possível avaliar uma diferença de comportamento no inicio da pandemia em relação ao restante dos gráficos;
* A presença de novas variantes do vírus no Reino Unido, Brasil e Africa do Sul potencializaram a sua contaminação a partir de dezembro de 2020;
* As datas festivas de final de ano podem ter contribuído para o aumento de casos e mortes;
* As vacinações podem ter minimizado o número de casos e mortes reportadas, conforme queda evidenciada no inicio de 2021. 



### **Q2. Quais são os 5 países com mais casos e mortes acumulados?**

Vamos avaliar os 5 países que mais possuem casos e mortes acumulados mundialmente. Para esta análise, avaliaremos os dados absolutos e os normalizados por população. 

É importante destacar que os dados absolutos tem maior influência em países mais populosos, já que o índice de propagação do vírus é maior. No entanto, para uma avaliação comparativa, é necessário normalizar os dados de tal forma que elimine o efeito da população, trazendo para a mesma base. Por isso, a importância de utilizar a métrica dos dados normalizados por população.

### **Q2. Quais são os 5 estados com mais casos e mortes acumuladas?**

Vamos avaliar os 5 estados que mais possuem casos e mortes acumuladas de forma similar a questão levantada para avaliação dos países na perspectiva mundial, utilizando as mesmas métricas.

```
#Definindo um dataframe com os 5 estados mais representativos em relação a casos acumulados
df_sel_cases = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'totalCases', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_cases = df_sel_cases['state'].to_list()

#Definindo uma variável para total de casos acumulados por um milhão
total_cases = df_sel_cases['totalCases'].apply(lambda x: x/1000000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a mortes acumulados
df_sel_deaths = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'deaths', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_deaths = df_sel_deaths['state'].to_list()

#Definindo uma variável para total de mortes acumulados por um milhão
total_deaths = df_sel_deaths['deaths'].apply(lambda x: x/1000000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a casos acumulados por 100 mil habitantes
df_sel_cases_100k = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'totalCases_per_100k_inhabitants', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_cases_normalizados = df_sel_cases_100k['state'].to_list()

#Definindo uma variável para total de casos acumulados normalizados pela população
total_cases_normalizados = df_sel_cases_100k['totalCases_per_100k_inhabitants'].apply(lambda x: x/100000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a mortes acumulados por 100 mil habitantes
df_sel_deaths_100k = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'deaths_per_100k_inhabitants', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_deaths_normalizados = df_sel_deaths_100k['state'].to_list()

#Definindo uma variável para total de mortes acumulados normalizado pela população
total_deaths_normalizados = df_sel_deaths_100k['deaths_per_100k_inhabitants'].apply(lambda x: x/100000).to_list()

#Definindo os parâmetros de style para o matplotlib
rc_params_bar = { 'axes.edgecolor':'#787878', 
                  'axes.titlecolor':'#787878', 
                  'xtick.color': '#787878',
                  'ytick.labelsize': 10,
                  'axes.spines.top':False, 
                  'axes.spines.right': False, 
                  'axes.spines.bottom': False, 
                  'axes.spines.left': False, 
                  'ytick.labelleft': False,
                  'xtick.bottom': False,
                  'ytick.left': False,
                  'axes.titleweight': 'bold',
                  'axes.titlesize': 10,
                  'axes.titlepad': 12}

#setando os parâmetros no matplotlib/seaborn
with plt.rc_context(rc_params_bar):

  #Instanciando a figure e axes
  fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,10))

  #-------------------------Gráfico total de casos acumulados ----------------
  
  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_cases, 
              y = total_cases, 
              palette="Blues_r",
              ax = ax[0,0])  
  
  #Definindo o título
  ax[0,0].set_title("Total de casos acumulados (por milhão) - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_cases):

    #Imprimindo o rótulo acima da barra
    ax[0,0].text(desl_x - 0.15 , 
            desl_y + 0.05 , 
            str(round(desl_y,1)) , 
            fontsize = 10, 
            color = '#787878')
    
  #-------------------------Gráfico total de casos normalizados ----------------
  
  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_cases_normalizados, 
              y = total_cases_normalizados, 
              palette="Blues_r",
              ax = ax[0,1])  
  
  #Definindo o título
  ax[0,1].set_title("Total de casos acumulados e normalizados por população - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_cases_normalizados):

    #Imprimindo o rótulo acima da barra
    ax[0,1].text(desl_x - 0.25, 
                 desl_y + 0.003, 
                 str(round(desl_y,4)) , 
                 fontsize = 10, 
                 color = '#787878')
  
  #-------------------------Gráfico total mortes acumualda----------------------

  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_deaths, 
              y = total_deaths, 
              palette="OrRd_r",
              ax = ax[1,0])  
  
  #Definindo o título
  ax[1,0].set_title("Total de mortes acumuladas (por milhão) - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_deaths):

    #Imprimindo o rótulo acima da barra
    ax[1,0].text(desl_x - 0.2, 
                 desl_y + 0.001, 
                 str(round(desl_y,2)) , 
                 fontsize = 10, 
                 color = '#787878')
  
   #-------------------------Gráfico total mortes normalizados-----------------

  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_deaths_normalizados, 
              y = total_deaths_normalizados, 
              palette="OrRd_r",
              ax = ax[1,1])  
  
  #Definindo o título
  ax[1,1].set_title("Total de mortes acumuladas e normalizadas por população - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_deaths_normalizados):

    #Imprimindo o rótulo acima da barra
    ax[1,1].text(desl_x - 0.25, 
                 desl_y + 0.00005, 
                 str(round(desl_y,4)) , 
                 fontsize = 10, 
                 color = '#787878')

  #Plotando o gráfico
  plt.show()
```

Como já mencionado, o número de casos e mortes tem uma forte relação com o tamanho da população, por isso, que é importante normalizar os dados para eliminar efeito do tamanho da população e permitir uma comparação.

Diante disso, podemos notar que Roraima é o estado que possui mais casos acumulados quando comparado com os demais estados. No entanto, Amazônas é o que possui mais mortes e que podem estar associados a presença da nova variante e do colapso que ocorreu em Manaus por falta de oxigência para combater o COVID-19.

Outro fato que chama bastante atenção, é o Rio de Janeiro não estar presente no top 5 de maior quantidade de casos, mas estar no de mortes. Isto pode ser resultado de uma maior taxa de mortalidade por COVID-19 em relação aos demais estados.

```
#Definindo um dataframe com os 5 estados mais representativos em relação a vacinação (1ª dose) acumulados
df_sel_vacinados = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'vaccinated', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_vacinados = df_sel_vacinados['state'].to_list()

#Definindo uma variável para total de vacinação (1ª dose) por milhão
total_vacinados = df_sel_vacinados['vaccinated'].apply(lambda x: x/1000000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a vacinação (2ª dose) acumulados
df_sel_vacinados_complet = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'vaccinated_second', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_vacinados_complet = df_sel_vacinados_complet['state'].to_list()

#Definindo uma variável para total de vacinação (2ª dose) acumulada por milhão
total_vacinados_complet = df_sel_vacinados_complet['vaccinated_second'].apply(lambda x: x/1000000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a vacinação (1ª dose) acumulada por 100 mil habitanes
df_sel_vacinados_100k = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL") ].sort_values(by = 'vaccinated_per_100k_inhabitants', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_vacinados_normalizados = df_sel_vacinados_100k['state'].to_list()

#Definindo uma variável para total de vacinações (1ª dose) acumulados normalizados pela população
total_vacinados_normalizados = df_sel_vacinados_100k['vaccinated_per_100k_inhabitants'].apply(lambda x: x/100000).to_list()

#Definindo um dataframe com os 5 estados mais representativos em relação a vacinações (2ª dose) acumuladas por 100 mil habitantes
df_sel_vacinados_complet_100k = df_br.loc[ (df_br['date'] == '2021-03-12') & (df_br['state'] != "TOTAL")  ].sort_values(by = 'vaccinated_second_per_100k_inhabitants', ascending = False)[:5].copy()

#Definindo uma variável para location - top5
location_total_vacinados_complet_normalizados = df_sel_vacinados_complet_100k['state'].to_list()

#Definindo uma variável para total de vacinações (2ª dose) acumulados normalizado pela população
total_vacinados_complet_normalizados = df_sel_vacinados_complet_100k['vaccinated_second_per_100k_inhabitants'].apply(lambda x: x/100000).to_list()

#Definindo os parâmetros de style para o matplotlib
rc_params_bar = { 'axes.edgecolor':'#787878', 
                  'axes.titlecolor':'#787878', 
                  'xtick.color': '#787878',
                  'ytick.labelsize': 10,
                  'axes.spines.top':False, 
                  'axes.spines.right': False, 
                  'axes.spines.bottom': False, 
                  'axes.spines.left': False, 
                  'ytick.labelleft': False,
                  'xtick.bottom': False,
                  'ytick.left': False,
                  'axes.titleweight': 'bold',
                  'axes.titlesize': 10,
                  'axes.titlepad': 12}

#setando os parâmetros no matplotlib/seaborn
with plt.rc_context(rc_params_bar):

  #Instanciando a figure e axes
  fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,10))

  #-------------------------Gráfico total de vacinados (1ª dose) acumulados ----------------
  
  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_vacinados, 
              y = total_vacinados, 
              palette="Greys_r",
              ax = ax[0,0])  
  
  #Definindo o título
  ax[0,0].set_title("Total de vacinações (1ª dose) acumulados (por milhão) - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_vacinados):

    #Imprimindo o rótulo acima da barra
    ax[0,0].text(desl_x - 0.15 , 
            desl_y + 0.005, 
            str(round(desl_y,1)) , 
            fontsize = 10, 
            color = '#787878')
    
  #-------------------------Gráfico total vacinados (1ª dose) normalizados -----
  
  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_vacinados_normalizados, 
              y = total_vacinados_normalizados, 
              palette="Greys_r",
              ax = ax[0,1])  
  
  #Definindo o título
  ax[0,1].set_title("Total de vacinações (1ª dose) e normalizados por população - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_vacinados_normalizados):

    #Imprimindo o rótulo acima da barra
    ax[0,1].text(desl_x - 0.15, 
                 desl_y + 0.0005, 
                 str(round(desl_y,2)) , 
                 fontsize = 10, 
                 color = '#787878')
  
  #-------------------------Gráfico total de vacinados completamente acumulados

  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_vacinados_complet, 
              y = total_vacinados_complet, 
              palette="Greys_r",
              ax = ax[1,0])  
  
  #Definindo o título
  ax[1,0].set_title("Total de vacinados completamente (por milhão) - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_vacinados_complet):

    #Imprimindo o rótulo acima da barra
    ax[1,0].text(desl_x - 0.1, 
                 desl_y + 0.008, 
                 str(round(desl_y,1)) , 
                 fontsize = 10, 
                 color = '#787878')
  
   #--------Gráfico total vacinados completamente acumulados e normalizados-----

  #Fazendo o gráfico de barras com o Seaborn
  sns.barplot(x = location_total_vacinados_complet_normalizados, 
              y = total_vacinados_complet_normalizados, 
              palette="Greys_r",
              ax = ax[1,1])  
  
  #Definindo o título
  ax[1,1].set_title("Total de vacinados completamente e normalizadas por população - top 5")

  #Colocando rótulo nos gráficos de barra
  for desl_x, desl_y in enumerate(total_vacinados_complet_normalizados):

    #Imprimindo o rótulo acima da barra
    ax[1,1].text(desl_x - 0.15, 
                 desl_y + 0.0001, 
                 str(round(desl_y,2)) , 
                 fontsize = 10, 
                 color = '#787878')

  #Exibindo o gráfico
  plt.show()
```
